#1) Monthly revenue and order volume (business growth over time)

###Shows how sales and order counts evolve over time (seasonality, growth, etc.).

Analysis: Looking at the results, you can immediately see a strong and consistent upward trend in both the number of delivered orders and the total GMV (gross merchandise value) from late 2016 through mid-2018.

The early months (late 2016 to early 2017) start with very low activity just a handful of orders and minimal revenue. But beginning around February 2017, order volume begins to pick up sharply. By mid-2017, the platform is already processing 3,000–4,000 orders per month, showing that customer acquisition and marketplace activity accelerated rapidly.

The biggest jump happens in the final quarter of 2017, where monthly order volume nearly doubles compared to mid-year. GMV follows the same pattern, climbing steadily from ~400k in early 2017 to well over 1 million per month in 2018.

By early 2018 the platform stabilizes at roughly 6,000–7,000 delivered orders per month, indicating a more mature and predictable sales cycle. Even though growth slows later in 2018, revenue remains consistently above the one-million-mark, demonstrating strong customer retention and continued demand.

```python
q1 = """SELECT strftime('%Y-%m', o.order_purchase_timestamp) AS year_month,
    COUNT(DISTINCT o.order_id) AS num_orders,
    ROUND(SUM(oi.price + oi.freight_value), 2) AS gmv -- gross merchandise value
FROM orders o
JOIN order_items oi ON o.order_id = oi.order_id
WHERE o.order_status = 'delivered'
GROUP BY year_month
ORDER BY year_month;
"""
out = pd.read_sql(q1, conn)
out
```

#2) Number of orders per product category

Analysis: This table highlights which product categories generate the highest number of sold items across the entire marketplace. The distribution is very uneven, with a few categories dominating sales while many niche categories contribute only a tiny fraction of total volume.

The fact that only a handful of categories dominate sales suggests a marketplace where demand is heavily concentrated in essential or everyday-use products.

```python
Q2 = """SELECT
    p.product_category_name,
    COUNT(*) AS items_sold
FROM order_items oi
JOIN products p ON oi.product_id = p.product_id
GROUP BY p.product_category_name
ORDER BY items_sold DESC;
"""
out = pd.read_sql(Q2, conn)
out
```

#3) Average review score by product category

Analysis: The results show clear differences in customer satisfaction across categories. Standardized, low-risk items like CDs/DVDs, children’s clothing, and books receive the highest reviews, likely because expectations are clear and products rarely arrive damaged. Lower-scoring categories such as hygiene products, kitchen appliances, and especially “seguros_e_servicos” indicate recurring issues with quality, packaging, or customer expectations. The large gap between the top and bottom categories highlights where the marketplace should investigate product quality and seller performance.

```python
Q3 = """SELECT
    p.product_category_name,
    ROUND(AVG(r.review_score), 2) AS avg_review
FROM order_reviews r
JOIN order_items oi ON r.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY p.product_category_name
ORDER BY avg_review DESC;
"""
out = pd.read_sql(Q3, conn)
out
```

#4) Total revenue per seller

# 8) Average review score per state

Analysis: Customer satisfaction varies across states, but overall review scores remain relatively high nationwide, with most states averaging above 4.0. Smaller states like Amapá (AP) and Paraná (PR) top the list, indicating particularly positive customer experiences. Meanwhile, larger markets such as Rio de Janeiro (RJ) and Bahia (BA) score noticeably lower, which may reflect delivery delays, product issues, or higher customer expectations. # 9) Seller performance: revenue, ratings, delivery time

Analysis: Among sellers with meaningful volume (50+ orders), the top performers combine strong review scores with reasonably fast delivery times. Several sellers from São Paulo (SP) show excellent ratings above 4.7, suggesting reliable fulfillment and high customer satisfaction. Sellers in other states, like Goiás (GO) and Rio Grande do Sul (RS), also appear in the top group but tend to have slightly longer delivery times. Overall, the results show that the highest-rated sellers are not necessarily the ones generating the most revenue, indicating opportunities to strengthen logistics or product mix for strong performers.

```python
Q9 = """SELECT
    s.seller_id,
    s.seller_state,
    COUNT(DISTINCT oi.order_id) AS num_orders,
    ROUND(SUM(oi.price), 2) AS revenue,
    ROUND(AVG(r.review_score), 2) AS avg_review_score,
    ROUND(
        AVG(
            julianday(o.order_delivered_customer_date)
          - julianday(o.order_purchase_timestamp)
        ),
        2
    ) AS avg_delivery_days
FROM sellers s
JOIN order_items oi
      ON s.seller_id = oi.seller_id
JOIN orders o
      ON oi.order_id = o.order_id
LEFT JOIN order_reviews r
      ON o.order_id = r.order_id
WHERE o.order_status = 'delivered'
GROUP BY s.seller_id, s.seller_state
HAVING num_orders >= 50          -- focus on meaningful volume
ORDER BY avg_review_score DESC, revenue DESC
LIMIT 10;
"""
out = pd.read_sql(Q9, conn)
out

```

# 10) Revenue and average order value by customer state

Analysis: São Paulo (SP) dominates total revenue by a huge margin, driven mainly by its extremely high order volume rather than high order values. Interestingly, several lower-volume states—such as Ceará (CE), Pará (PA), and Maranhão (MA)—show much higher average order values, often above R$200, indicating that customers in these regions spend more per purchase. This suggests that while SP is the largest market, smaller states may hold high-value customer segments worth targeting with premium products. The wide variation in order values highlights meaningful differences in regional purchasing power and shopping behavior.

```python
Q10 = """SELECT
    customer_state,
    COUNT(*) AS num_orders,
    ROUND(SUM(order_value), 2) AS total_revenue,
    ROUND(AVG(order_value), 2) AS avg_order_value
FROM (
    SELECT
        o.order_id,
        c.customer_state,
        SUM(oi.price + oi.freight_value) AS order_value
    FROM orders o
    JOIN customers c
          ON o.customer_id = c.customer_id
    JOIN order_items oi
          ON oi.order_id = o.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY o.order_id, c.customer_state
) t
GROUP BY customer_state
ORDER BY total_revenue DESC;
"""
out = pd.read_sql(Q10, conn)
out
```

# 13) Payment value vs review score

Analysis: Lower review scores (1–2 stars) are associated with noticeably higher average payment values, suggesting that expensive orders may create higher customer expectations—and greater disappointment when something goes wrong. As review scores increase from 3 to 5, average payment values stay relatively stable and lower, which implies that smaller purchases tend to generate fewer extreme negative experiences. This trend hints at a possible “price sensitivity effect,” where customers who spend more are more critical when the buying experience isn’t perfect. The insight can help the business prioritize support and quality control for high-value orders.

```python
Q13 = """SELECT
    r.review_score,
    ROUND(AVG(op.payment_value), 2) AS avg_payment
FROM order_reviews r
JOIN orders o
    ON r.order_id = o.order_id
JOIN order_payments op
    ON o.order_id = op.order_id
GROUP BY r.review_score
ORDER BY r.review_score;
"""
out = pd.read_sql(Q13, conn)
out
```

# 14) Customer states with the highest number of unique product categories purchased

Analysis: Customers from São Paulo (SP), Rio de Janeiro (RJ), and Minas Gerais (MG) purchase the widest variety of product categories, with each state showing 70+ unique categories—a sign of large, diverse consumer markets. Southern states like RS and SC also show strong variety, reinforcing their role as mature e-commerce regions with broad shopping behavior. Smaller states such as PI, MA, and PA still exhibit meaningful category diversity, though at lower levels, likely reflecting smaller populations and thinner product demand. # 15) Average shipping cost per product category

Analysis: Categories like computers, large home appliances, and furniture have the highest shipping costs, all averaging above R40, which makes sense given their heavy weight and bulky dimensions. These high freight costs can significantly influence purchase decisions, especially for price-sensitive customers. At the other end of the spectrum, small and lightweight items—such as fashion_children’s_clothes, books_imported, and female_clothing—have much lower shipping fees, often below R$13. The wide cost range highlights the need for differentiated logistics strategies, especially for categories where freight may be a barrier to conversion.

```python
Q15 = """SELECT
    t.product_category_name_english AS category,
    ROUND(AVG(oi.freight_value), 2) AS avg_freight
FROM order_items oi
JOIN products p
    ON oi.product_id = p.product_id
JOIN product_category_name_translation t
    ON p.product_category_name = t.product_category_name
GROUP BY category
ORDER BY avg_freight DESC;
"""

out = pd.read_sql(Q15, conn)
out
```

###Some product categories are extremely expensive to ship — more than their price.

Some categories (especially bulky or low-value items) cost more to ship than the product itself. This means the company may be losing margin or relying heavily on customer shipping fees.

Analysis: Some categories—like home_comfort_2, flowers, and diapers_and_hygiene—show freight costs that are very close to the item price, meaning shipping represents a disproportionate share of the total cost. This can seriously squeeze margins if the company subsidizes shipping, or hurt conversion rates if customers must pay these high fees. Meanwhile, high-value categories such as computers or home appliances still have high freight, but the cost becomes negligible relative to the price. These findings suggest the business should revisit shipping strategies for low-price, bulky items to avoid unprofitable logistics.

```python
Q16 = """SELECT
    t.product_category_name_english AS category,
    ROUND(AVG(oi.freight_value), 2) AS avg_freight,
    ROUND(AVG(oi.price), 2) AS avg_price,
    ROUND(AVG(oi.freight_value) - AVG(oi.price), 2) AS freight_minus_price
FROM order_items oi
JOIN products p
    ON oi.product_id = p.product_id
JOIN product_category_name_translation t
    ON p.product_category_name = t.product_category_name
GROUP BY category
ORDER BY freight_minus_price DESC;
"""
out = pd.read_sql(Q16, conn)
out
```

###Customers in northern states wait WAY longer for deliveries

Analysis: Customers in northern states like Roraima (RR), Amapá (AP), and Amazonas (AM) face dramatically longer delivery times—often 25–30 days, nearly double the national average. These delays likely stem from long transport distances, limited carrier availability, and weaker logistics infrastructure in the North. Meanwhile, states in the South and Southeast experience much faster delivery, reflecting stronger transportation networks and higher-density hubs. This gap highlights a major opportunity for logistics optimization or regional fulfillment strategies to improve customer experience in under-served areas.

```python
Q17 = """ SELECT
    c.customer_state,
    ROUND(AVG(julianday(o.order_delivered_customer_date) - julianday(o.order_purchase_timestamp)), 2) AS avg_delivery_days
FROM orders o
JOIN customers c
    ON o.customer_id = c.customer_id
WHERE o.order_status = 'delivered'
GROUP BY c.customer_state
ORDER BY avg_delivery_days DESC;
"""
out = pd.read_sql(Q17, conn)
out
```
